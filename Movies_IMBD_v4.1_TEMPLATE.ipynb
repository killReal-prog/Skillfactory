# Предобработка

```python
answers = {} # создадим словарь для ответов

#преобразуем дату в объект Date
def to_date(date_string):
    date_list = date_string.split('/')
    date=datetime.date(int(date_list[2]),int(date_list[0]),int(date_list[1]))
    return date

source_df.release_date=source_df.release_date.apply(to_date)

#преобразуем строки с перечислением объектов в списки
source_df.cast=source_df.cast.apply(lambda s: s.split('|'))
source_df.genres=source_df.genres.apply(lambda s: s.split('|'))
source_df.production_companies=source_df.production_companies.apply(lambda s: s.split('|'))
source_df.director=source_df.director.apply(lambda s: s.split('|'))

#добавим столбец profit
source_df['profit']=source_df['revenue']-source_df['budget']
```

# 13. У какого режиссера самые большие суммарные кассовые сборы?

```python
'''Идея в том чтобы разбить записи у которых несколько режиссеров на несколько записей и применить groupby
Если можно было это реализовать эффективней - дайте знать'''
data = source_df.copy()

temp_df=pd.DataFrame(columns=['director', 'revenue'])

for i in data.index:
    for director in data.director[i]:
        new_row = {'director':director, 'revenue':data.revenue[i]}
        temp_df = temp_df.append(new_row, ignore_index=True)
        
temp_df.groupby('director')['revenue'].sum().sort_values()

answers['13']='Peter Jackson'
```

# 14. Какой режисер снял больше всего фильмов в стиле Action?

```python
data = source_df.copy()

temp_df=pd.DataFrame(columns=['director', 'genres'])

#снова разделяем записи так чтобы был один режиссер
for i in data.index:
    for director in data.director[i]:
        new_row = {'director':director, 'genres':data.genres[i]}
        temp_df = temp_df.append(new_row, ignore_index=True)

def is_action(genres):
    if 'Action' in genres:
        return True
        
temp_df['is_action']=temp_df.genres.apply(is_action)
#можно было сделать это проще или например lambda функцией?

temp_df.groupby('director')['is_action'].count().sort_values()

answers['14']='Robert Rodriguez'

#вам как профессионалу вообще понятно что я делаю или надо больше комментариев?
```

# 15. Фильмы с каким актером принесли самые высокие кассовые сборы в 2012 году? 

```python
data = source_df.copy()

temp_df=pd.DataFrame(columns=['actor', 'film_revenue'])

#снова разделяем записи так чтобы был один актер
for i in data.index:
    if data.release_year[i]==2012:
        for actor in data.cast[i]:
            new_row = {'actor':actor, 'film_revenue':data.revenue[i]}
            temp_df = temp_df.append(new_row, ignore_index=True)
        
#дальше решение зависит от того как трактовать вопрос то есть какую агрегатную функцию применить
temp_df = temp_df.astype({'film_revenue': np.int})
temp_df.groupby('actor')['film_revenue'].sum().sort_values()

answers['15']='Chris Hemsworth '
```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
data = source_df['cast'].copy()

data1 = data.explode('cast')

actors = data1.unique()

a_df = pd.DataFrame(0, columns=actors, index=actors)#таблица пересечения всех актеров со всеми

for actors in data:
    for couple in combinations(actors, 2):
        a_df[couple[0]][couple[1]]+=1

max_cast = 0

for i in a_df.columns:
    for j in a_df.index:
        if a_df[i][j]>max_cast:
            max_cast = a_df[i][j]

#т.к. таких пар несколько пришлось использовать еще один цикл
#можно было избежать этого?
for i in a_df.columns:
    for j in a_df.index:
        if a_df[i][j]==max_cast:
            print(i,j)

#код выполняется минуты полторы. Является ли он оптимальным?
```
